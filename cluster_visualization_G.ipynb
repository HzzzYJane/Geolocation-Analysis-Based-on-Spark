```python
# cluster visualzie #Euclidean k=2
latitude=40.428
longitude=-74.1
dir_G2 = glob.glob('/Users/hanzheyu/Desktop/CSE427-project/NY_pickup/result2G/2G/points/*.txt')
df = pd.concat([pd.read_csv(f, sep=" ", header=None) for f in dir_G2])
df.rename({0:'classify', 1: 'lat', 2: 'lon'})
df.columns = ['classify','lat', 'lon']
# df = pd.read_csv(r'C:\Users\zhang\Desktop\ k-means\k=3\part-00001.csv',index_col=0)
df = df.iloc[1:25000,[0,1,2]]

df0 = df[df.classify==0]

df1 = df[df.classify==1]

incidents=folium.map.FeatureGroup()

for lat,lon, in zip(df0.lat,df0.lon):
        incidents.add_child(
            folium.CircleMarker(
            [lat,lon],
            radius=0.1,
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=4
            )
        
        )
for lat,lon, in zip(df1.lat,df1.lon):
        incidents.add_child(
            folium.CircleMarker(
            [lat,lon],
            radius=0.1,
            color='green',
            fill=True,
            fill_color='green',
            fill_opacity=4
            )     
        )

# m = folium.Map(location=[39.712183, -104.998424], zoom_start=5)
# point_layer = folium.FeatureGroup(name="Query Search")


ny_map=folium.Map(location=[latitude,longitude],zoom_start=12)
ny_map.add_child(incidents)
point_layer = folium.FeatureGroup(name="Query Search")

latLong = [(40.7228350701,-73.9662663023,"first cluster"),
           (40.7622940261,-73.9779599513,"second cluster")]

for lat,lng,nameP in latLong:
    point_layer.add_child(folium.CircleMarker(location=[lat, lng], radius=2,
        popup=str(nameP) + " Lat: " + str(lat) + " , Long: " + str(lng), 
        tooltip=str(nameP) + " Lat: " + str(lat) + " , Long: " + str(lng),
        fill=True,  # Set fill to True
        color='red',
        fill_opacity=1)).add_to(ny_map)
    
ny_map.add_child(point_layer)
ny_map.add_child(folium.LayerControl()) 
    
ny_map.save("/Users/hanzheyu/Desktop/CSE427-project/screenshot/mymapG2.html")
```
